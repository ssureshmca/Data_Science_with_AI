```python


file_path='emails.txt'
file=open(file_path,encoding='utf-8-sig')

file_list=[line.split('; ') for line in file]

name_list=[i.split('<')[0].split(',')[0] for i in file_list[0]]
print(name_list)

surname_list=[i.split('<')[0].split(',')[1] for i in file_list[0]]
print(surname_list)

company_list=[i.split('<')[1].rstrip('>').split('@')[1].rstrip('.com') for i in file_list[0]]
print(company_list)

emailid_list=[i.split('<')[1].rstrip('>') for i in file_list[0]]
print(emailid_list)
```
