```python
dtypes = dict(churn_df.dtypes)
Category_columns=[i for i in dtypes if dtypes[i]=='O']
churn_df[Category_columns]
```

```python
from sklearn.preprocessing import LabelEncoder  # Import the package
le=LabelEncoder()  

for i in Category_columns[1:]:
    churn_df[i]=le.fit_transform(churn_df[i]) # Apply fit transform
    
churn_df    
```
