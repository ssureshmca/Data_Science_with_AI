```python
import numpy as np
import matplotlib.pyplot as plt
```

$Step - 2:$

**Read the data**

```python
data=np.random.exponential(size=100000)
# we are taking a random values
# 1000 sample we are taking
```

```python
plt.hist(data,bins=40, label='Exponential')
plt.legend()
plt.show()
```

$Step - 3:$

**Log Transformation**

- np.log represents natural logoritham

- natural logorithm means base e

- exponential will multiply with log base e

- 

```python
log_data=np.log(data)
plt.hist(log_data,bins=40, label='Log transformation')
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1).hist(data,bins=40)
plt.subplot(1,2,2).hist(log_data,bins=40)
plt.show()
```

$Step - 4:$

**Reciprocal transformation**

```python
rec_data=1/data
plt.hist(rec_data,bins=40, label='Reciprocal transformation')
plt.legend()
plt.show()
```
