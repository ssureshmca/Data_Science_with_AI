```python
# Step-1: Calculate min value of P-wage
# Step-2: Calculate max value of p-wage
# Step-3: Dr=max_wage-min_wage
# Step-4: Nr=P_wage-min_wage
# Step-5: Nr/Dr

pwage_min=visa_df['prevailing_wage'].min()
pwage_max=visa_df['prevailing_wage'].max()
dr=pwage_max-pwage_min
nr=visa_df['prevailing_wage']-pwage_min
visa_df['prevailing_wage_Norm']=nr/dr
visa_df[['prevailing_wage','prevailing_wage_Norm']]
```

```python
visa_df['prevailing_wage_Norm'].min()
```

```python
max_id=visa_df['prevailing_wage_Norm'].idxmax()
min_id=visa_df['prevailing_wage_Norm'].idxmin()

max_id,min_id
```

```python
visa_df[['prevailing_wage','prevailing_wage_Norm']].iloc[[max_id,min_id]]
```

```python
############## Read the data again ########################
fileName = "Visadataset.csv"
visa_df=pd.read_csv(fileName)

#Step-1
from sklearn.preprocessing import MinMaxScaler

#Step-2
mms=MinMaxScaler()

#Step-3
visa_df['prevailing_wage_norm1']=mms.fit_transform(visa_df[['prevailing_wage']])
```

**Standardization**

```python
p_mean=visa_df['prevailing_wage'].mean()
p_std=visa_df['prevailing_wage'].std()
visa_df['prevailing_wage_zscore']=(visa_df['prevailing_wage']-p_mean)/p_std
visa_df[['prevailing_wage','prevailing_wage_zscore']]
```

```python
z_min=visa_df['prevailing_wage_zscore'].min()
z_max=visa_df['prevailing_wage_zscore'].max()
z_min,z_max
```
